```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data[[0,1]],data[2],test_size=0.1)
```

```python
import tensorflow 
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_trainf=std.fit_transform(x_train)
x_testf=std.transform(x_test)
```

```python
model = Sequential()
model.add(Dense(6,activation="relu",input_dim=2))
model.add(Dense(6,activation="relu"))
model.add(Dense(5,activation="relu"))
model.add(Dense(4,activation="relu"))
model.add(Dense(3,activation="relu"))
model.add(Dense(2,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history=model.fit(x_trainf,y_train,batch_size=10,epochs=100,validation_split=0.2)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
```

```python
y_cap=model.predict(x_testf)
```

```python
y_predf = np.array([1 if x>0.5 else 0 for x in y_cap])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_capf)
```

```python
from mlxtend.plotting import plot_decision_regions as pt
```

```python
plt.figure(figsize=(8,6),dpi=150)
pt(x_train.values.astype(int),y_train.values.astype(int),clf=model)
plt.title("X_train Decision regions",color='red',fontsize='medium',fontfamily='serif')
plt.xlabel("X_train",color='red',fontsize='medium',fontfamily='serif')
plt.ylabel("y_train",color='red',fontsize='medium',fontfamily='serif')
plt.show()
```

```python
plt.figure(figsize=(8,6),dpi=150)
pt(x_test.values,y_predf,clf=model)
plt.title("X_test Decision regions",color='red',fontsize='medium',fontfamily='serif')
plt.xlabel("X_test",color='red',fontsize='medium',fontfamily='serif')
plt.ylabel("y_test",color='red',fontsize='medium',fontfamily='serif')
plt.show()
```
